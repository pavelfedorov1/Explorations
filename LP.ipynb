### Linear Programming Example

### Primal Problem

$$max[ 2x_1+3x_2]$$

$$3x1+ x2 \le 2$$

$$x1+3x2 \le 5$$

$$âˆ’ x1+2x2 \le 3$$

$$x\ge 0$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of values for x
x = np.linspace(0, 1, 400)

# Define the lines
y1 = (2 - 3*x)
y2 = (5 - x) / 3
y3 = (x + 3) / 2
y4 = (5 - 2*x) / 3  # New line

# Create the plot
plt.figure()

# Plot the lines
plt.plot(x, y1, label="3x1 + x2 <= 2")
plt.plot(x, y2, label="x1 + 3x2 <= 5")
plt.plot(x, y3, label="-x1 + 2x2 <= 3")
plt.plot(x, y4, label="max(z) = 5 = 2x1 + 3x2", linestyle='dotted')

# Fill the feasible region
y5 = np.minimum(y1, y2)
y6 = np.minimum(y5, y3)
plt.fill_between(x, 0, y6, where=(y6>=0), color='gray', alpha=0.5)

plt.xlabel('x1')
plt.ylabel('x2')
optimum_x = 1/7
optimum_y = 11/7

plt.scatter(optimum_x, optimum_y, marker='o', c="r", s=100, label='solution')
plt.annotate(f'optimum (1/7, 11/7)', (1/7, 11/7), textcoords="offset points",
             xytext=(10,10), ha='center',c='r') 

plt.title('Feasible region for the given system of inequalities')
plt.legend()
plt.grid(True)
plt.xlim(0, 1)
plt.ylim(0, 2)
plt.show()

print('note that constraint #2 is not binding')
```

### Dual Problem 

$$min (2,5,3)y $$

$$3y_1+y_2-y_3 = 2$$

$$y_1+3y_2+2y_3 = 3$$


$$y\ge0$$

Solving the Primal using Fourier Motzkin (FM) we can obtain the dual solution $y = (0,1,0)^{T}$

We follow the linear combinations of inequalities to obtain the equality that yields the optimum using FM.

$$min (2,5,3)(0,1,0)^{T} = 5$$

Both the primal and the dual have feasible solutions => strong duality holds and primal max = dual min

We can also confirm strong duality using the complementary slackness theorem that states that for each value in the dual solution y

$$y_i(a_i^Tx^{*}-b_i) = 0$$

so lets calculate

$$1(3 \frac{1}{7}+\frac{11}{7}-2) = 0$$

$$0(\frac{1}{7}+3\frac{11}{7}-5) = 0$$

$$1(-\frac{1}{7}+2\frac{11}{7}-3) = 0$$

Hence this is another way to confirm strong duality holds for this LP
